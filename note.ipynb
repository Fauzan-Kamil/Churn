```python
# membuat Churn label
# referensi: https://www.linkedin.com/pulse/customer-Churn-analysis-oversimplified-wrong-ian-pollard/
# Mengubah kolom 'last_purchase_date' menjadi datetime
df['last_purchase_date'] = pd.to_datetime(df['last_purchase_date'])

# Menentukan periode waktu untuk memeriksa Churn
churn_period = pd.Timedelta(days=180)

# Menghitung tanggal terakhir dalam dataset
max_date = df['last_purchase_date'].max()

Churn = []
# Looping untuk setiap customer
for i, row in df.groupby('customer_id').last().iterrows():
    last_purchase_date = pd.to_datetime(row['last_purchase_date'])
    diff = max_date - last_purchase_date
    if row['Customer_segment'] == "Lost Customers":
        if diff >= churn_period:
            Churn.append(1)
        else:
            Churn.append(0)
    elif row['Customer_segment'] == "At-Risk Customers":
        if diff >= churn_period:
            Churn.append(1)
        else:
            Churn.append(0)
    elif row['Customer_segment'] == "Promising Customers":
        if diff >= churn_period:
            Churn.append(1)
        else:
            Churn.append(0)
    elif row['Customer_segment'] == "Loyal Customers":
        if diff >= churn_period:
            Churn.append(1)
        else:
            Churn.append(0)
    elif row['Customer_segment'] == "Best Customers":
        if diff >= churn_period:
            Churn.append(1)
        else:
            Churn.append(0)

pd_churn = pd.DataFrame(Churn, columns = ['Churn'])
df['Churn'] = pd_churn
```
